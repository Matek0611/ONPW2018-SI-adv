```python
import numpy as np
import pandas as pd
```

```python
planes = pd.read_csv('../nycflights13_planes.csv.gz', comment = '#')
```

```python
planes.head()
```

```python
planes.engine.drop_duplicates()
```

```python
planes.loc[:,['engine', 'type']].drop_duplicates().reset_index(drop=True)
```

```python
planes.groupby('engine').size().reset_index()
```

```python
planes.groupby(['engine', 'type']).size().reset_index()
```

```python
planes.query('100 < seats < 200').reset_index(drop=True).head()
```

```python
planes.query('(100 < seats < 200) and (year >= 2010)').reset_index(drop=True).head()
```

```python
planes.query('(seats >= 300) and (manufacturer in ["BOEING", "AIRBUS", "EMBRAER"])').reset_index(drop=True).head()
```

```python
planes.query('seats >= 200').groupby('manufacturer').size().reset_index()
```

```python
planes.groupby('manufacturer').size().to_frame('number').reset_index()
```

```python
planes.query('seats >= 200').groupby('manufacturer').size().to_frame('number').reset_index().query('number >= 10')
```
